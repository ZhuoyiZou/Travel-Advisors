```python
def get_coords(coord_string):
    if pd.notna(coord_string):
        temp = re.findall(r"(?<=\().+(?=\))",coord_string)
        if len(temp) == 2:
            lat,lng = temp[1].split(",")
        else:
            lat,lng = temp[0].split(",")
    else:
        lat,lng = [None,None]
    return([lat,lng])
```

```python
renamed_df_copy["Origin Lat"] = renamed_df_copy["Origin Coordinates"].map(lambda x: get_coords(x)[0])
renamed_df_copy["Origin Lng"] = renamed_df_copy["Origin Coordinates"].map(lambda x: get_coords(x)[1])
renamed_df_copy["Destination Lat"] = renamed_df_copy["Origin Coordinates"].map(lambda x: get_coords(x)[0])
renamed_df_copy["Destination Lng"] = renamed_df_copy["Origin Coordinates"].map(lambda x: get_coords(x)[1])
```
